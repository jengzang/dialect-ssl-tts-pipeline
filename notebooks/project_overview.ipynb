## 常见问题

### Q1: 如何选择合适的模型？
- **小数据集 (<1000 样本)**: 使用 SVM 或传统机器学习
- **中等数据集 (1000-10000 样本)**: 使用 LSTM 或小型神经网络
- **大数据集 (>10000 样本)**: 使用预训练模型微调（wav2vec, LoRA）

### Q2: GPU 显存不足怎么办？
- 减小 batch_size
- 使用梯度累积
- 使用混合精度训练
- 使用量化（4-bit, 8-bit）

### Q3: 如何提高模型性能？
- 增加训练数据
- 数据增强
- 调整超参数
- 使用集成学习
- 尝试不同的模型架构

### Q4: 如何部署模型？
- 导出为 ONNX 格式
- 使用 TorchScript
- 量化模型以减小体积
- 使用 FastAPI 构建 API 服务
